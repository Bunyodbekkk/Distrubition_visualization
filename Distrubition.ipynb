```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/blob/main/uybor_scrapping.csv?raw=true")
df.head()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(data=df,x='price')
plt.show()
```

```python
df=df[df['price']<200000]
plt.figure(figsize=(10,6))
sns.histplot(data=df,x='price')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(data=df,x='size')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(data=df,x='rooms')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(data=df,x='price',hue='rooms')
plt.show()
```

```python
a=df[df['rooms']<4]
plt.figure(figsize=(10,6))
sns.histplot(data=a,x='price',hue='rooms')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.displot(data=df[df['rooms']>3],x='price',col="rooms")
plt.show()
```

```python
sns.displot(data=df,x='size',kind='kde',height=5,aspect=2)
```

```python
sns.displot(data=df,x='size',kde=True,height=5,aspect=2)
```

```python
c=df[df['rooms'].isin([2,3,5])]
sns.displot(data=c,x='price',hue='rooms',kind='kde',height=5,aspect=2)
```

```python
sns.displot(data=df,x='size',kind='ecdf',height=5,aspect=2)
plt.grid(which='both')
plt.minorticks_on()
```

```python
sns.displot(data=df,x='price',kind='ecdf',height=5,aspect=2)
plt.grid(which='both')
plt.minorticks_on()
```
